```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

<h2>Reading eyes</h2>

```python
image_1 = cv2.imread("./Resources/1.bmp", cv2.IMREAD_GRAYSCALE)
image_2 = cv2.imread("./Resources/2.bmp", cv2.IMREAD_GRAYSCALE)
image_3 = cv2.imread("./Resources/3.bmp", cv2.IMREAD_GRAYSCALE)
image_4 = cv2.imread("./Resources/4.bmp", cv2.IMREAD_GRAYSCALE)
image_5 = cv2.imread("./Resources/5.bmp", cv2.IMREAD_GRAYSCALE)

plt.imshow(image_1)
```

<h2>Image 1</h2>

```python
blurred = cv2.medianBlur(image_1, 5)
edges = cv2.Canny(blurred , 10, 30) 
cimg = cv2.cvtColor(image_1,cv2.COLOR_GRAY2BGR)


# Use the Hough Circle Transform to detect circles
circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, 1, 14, param1=45, param2=71, minRadius=0, maxRadius=200)
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
    # Draw the outer circle
    cv2.circle(cimg, (i[0], i[1]), i[2], (0, 255, 0), 3)
    # Draw the center of the circle
    cv2.circle(cimg, (i[0], i[1]), 2, (255,0, 0), 2)

plt.figure(figsize=(18, 6))
plt.subplot(131), plt.imshow(image_1, cmap='gray'), plt.title('Original Image_1'), plt.axis('off')
plt.subplot(132), plt.imshow(edges, cmap='gray'), plt.title('Canny Edge Imag_1'), plt.axis('off')
plt.subplot(133), plt.imshow(cimg), plt.title('Detected Circles of Image_1'), plt.axis('off')
plt.show()

# Save image
cv2.imwrite("./Edge_Results/1_edge.bmp", edges)
cv2.imwrite("./Segmented_Results/1_segmented.bmp", cimg)
```

<h2>Helper tuning to work with different parameters</h2>

```python
# Display the window
cv2.namedWindow('Canny Edge Detection')

# Create trackbars for lower and upper thresholds
cv2.createTrackbar('Lower', 'Canny Edge Detection', 15, 100, nothing)
cv2.createTrackbar('Upper', 'Canny Edge Detection', 40, 100, nothing)

while True:
    # Get the current positions of trackbars
    lower = cv2.getTrackbarPos('Lower', 'Canny Edge Detection')
    upper = cv2.getTrackbarPos('Upper', 'Canny Edge Detection')

    blurred = cv2.medianBlur(image_4,7) # change images here
    edges = cv2.Canny(blurred, lower, upper)

    # Display the image
    cv2.imshow('Canny Edge Detection', edges)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
blurred = cv2.medianBlur((image_4), 13)
# Apply Canny edge detection
edges = cv2.Canny(blurred , 15 , 20)  
plt.imshow(edges)
```

```python
# Initial parameters
param1 = 50
param2 = 74
minRadius = 0
maxRadius = 0
dist = 10

cv2.namedWindow('Circle Detection Tuner', cv2.WINDOW_NORMAL)

while True:
    # Clone the original image
    img_clone = cv2.cvtColor(image_4,cv2.COLOR_GRAY2BGR)  # change images here
    
    # Detect circles. Now using 'edges' instead of 'blurred'.
    circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, 1,8, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)
     
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0, :]:
            cv2.circle(img_clone, (i[0], i[1]), i[2], (0, 255, 0), 3)
            cv2.circle(img_clone, (i[0], i[1]), 2, (0, 0, 255), 4)

    # Display the parameters on the side of the image
    cv2.putText(img_clone, f'param1: {param1}', (10, image_2.shape[0]-100), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    cv2.putText(img_clone, f'param2: {param2}', (10, image_2.shape[0]-75), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    cv2.putText(img_clone, f'minRadius: {minRadius}', (10, image_2.shape[0]-50), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    cv2.putText(img_clone, f'maxRadius: {maxRadius}', (10, image_2.shape[0]-25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    
    cv2.imshow('Circle Detection Tuner', img_clone)
    
    k = cv2.waitKey(1) & 0xFF
    
    if k == 27:  # ESC key to exit
        break
    elif k == ord('j'):  # 'J' key for decreasing param1
        param1 -= 1
    elif k == ord('l'):  # 'K' key for increasing param1
        param1 += 1
    elif k == ord('w'):  # 'L' key for decreasing param2
        param2 -= 1
    elif k == ord('s'):  # 'I' key for increasing param2
        param2 += 1
    elif k == ord('a'):  # 'A' key for decreasing minRadius
        minRadius -= 1
    elif k == ord('d'):  # 'D' key for increasing minRadius
        minRadius += 1
    # elif k == ord('w'):  # 'W' key for increasing maxRadius
    #     maxRadius += 1
    # elif k == ord('s'):  # 'S' key for decreasing maxRadius
    #     maxRadius -= 1
    elif k == ord('q'):  # 'S' key for decreasing maxRadius
        dist += 1

cv2.destroyAllWindows()
```
