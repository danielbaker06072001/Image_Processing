```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

<h1>Question 1: Feature detection Basics</h1>

```python
image1 = cv2.imread("./img/IMG_7099.jpg")
image2 = cv2.imread("./img/IMG_7100.jpg")

image1_gray = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

plt.imshow(image1)
```

```python
# Assuming 'image1_gray' and 'image2_gray' are your grayscale images
# and 'kp1' are the keypoints detected by SIFT for 'image1_gray'
# First, you need to create a SIFT object and detect keypoints for 'image2_gray'

sift_image1 = cv2.SIFT_create()
sift_image2 = cv2.SIFT_create()

kp1 = sift_image2.detect(image1_gray, None)
kp2 = sift_image2.detect(image2_gray, None)

# Convert the image to color so that you can display the keypoints in color
image1_color = cv2.cvtColor(image1_gray, cv2.COLOR_GRAY2BGR)
image2_color = cv2.cvtColor(image2_gray, cv2.COLOR_GRAY2BGR)

# Define the color and size of the keypoints
keypoint_color = (0, 255, 0)  # Green color
keypoint_size = 15  # The radius of the keypoint
keypoint_thickness = 3  # The thickness of the outline of the keypoint

# Draw each keypoint with the specified properties on image1
for kp in kp1:
    x, y = kp.pt
    cv2.circle(image1_color, (int(x), (int(y)), keypoint_size, keypoint_color, keypoint_thickness)

# You need to do the same for image2's keypoints
for kp in kp2:
    x, y = kp.pt
    cv2.circle(image2_color, (int(x), (int(y)), keypoint_size, keypoint_color, keypoint_thickness)

# Now you can display both images with keypoints highlighted
plt.figure(figsize=(20,10))

# Display the first image
plt.subplot(1, 2, 1)
plt.title('Image 1 Keypoints')
plt.imshow(cv2.cvtColor(image1_color, cv2.COLOR_BGR2RGB))

# Display the second image
plt.subplot(1, 2, 2)
plt.title('Image 2 Keypoints')
plt.imshow(cv2.cvtColor(image2_color, cv2.COLOR_BGR2RGB))

plt.show()
```
